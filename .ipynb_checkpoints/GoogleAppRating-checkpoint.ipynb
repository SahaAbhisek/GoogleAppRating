```python
data4 = pd.concat([data4, pd.get_dummies(data4['Category'], drop_first=True)], axis=1)
data4.drop(['Category'], axis=1, inplace=True)
```

```python
data4.shape, data3.shape
```

```python
data4 = pd.concat([data4, pd.get_dummies(data4['Content Rating'], drop_first=True)], axis=1)
data4.drop(['Content Rating'], axis=1, inplace=True)
```

```python
data4.head()
```

```python
data5 = data4.copy()
```

```python
all_genre = set()
for rows, index in data5.iterrows():
    app_genre = index[5].split(';')
    for genre in app_genre:
        all_genre.add(genre)
    
all_genre = pd.Series(list(all_genre))
for genre in all_genre:
    data5[genre] = 0
    
for rows, index in data5.iterrows():
    app = index[5].split(';')
    for genre in app:
        data5.at[rows, genre] = 1
        
data5.drop(['Genres'], axis=1, inplace=True)
```

### Making Prediction

```python
pip install sklearn
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# Splitting Data
X = data5.drop("Rating", axis=1)
Y = data5["Rating"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=132)

RF = RandomForestRegressor(n_estimators=20)
RF.fit(X_train, Y_train)
```

```python
Y_pred = RF.predict(X_test)
```

```python
#plt.figure(figsize=(6,4))
fig, ax = plt.subplots(1,2)
fig.set_size_inches(14,6)

ax[0].scatter(Y_test, Y_pred)
ax[0].set_xlabel(" True Rating", fontdict={'fontsize':14, 'fontweight':'bold'})
ax[0].set_ylabel(" Predicted Rating", fontdict={'fontsize':14, 'fontweight':'bold'})


residual = Y_test-Y_pred
sns.distplot(residual, ax=ax[1], color='g')
ax[1].set_xlabel("Residual", fontdict={'fontsize':14, 'fontweight':'bold'})
ax[1].set_ylabel("Density",fontdict={'fontsize':14, 'fontweight':'bold'})
plt.show()
```

```python
sns.regplot(x=Y_pred, y= residual, scatter=True)
```

```python
rf_mse = mean_squared_error(Y_test, Y_pred)
rf_rmse = np.sqrt(rf_mse)
print("Random Forest RMSE:   ",rf_rmse)

rf_mae = mean_absolute_error(Y_test, Y_pred)
print("Random Forest MAE :   ", rf_mae)
```

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
knn.fit(X_train, Y_train)
```

```python
knn_pred = knn.predict(X_test)
```

```python
#plt.figure(figsize=(6,4))
fig, ax = plt.subplots(1,2)
fig.set_size_inches(14,6)

ax[0].scatter(Y_test, knn_pred)
ax[0].set_xlabel(" True Rating", fontdict={'fontsize':14, 'fontweight':'bold'})
ax[0].set_ylabel(" Predicted Rating", fontdict={'fontsize':14, 'fontweight':'bold'})


residual = Y_test-knn_pred
sns.distplot(residual, ax=ax[1], color='g')
ax[1].set_xlabel("Residual", fontdict={'fontsize':14, 'fontweight':'bold'})
ax[1].set_ylabel("Density",fontdict={'fontsize':14, 'fontweight':'bold'})
plt.show()
```

```python
knn_mse = mean_squared_error(Y_test, knn_pred)
knn_rmse = np.sqrt(knn_mse)
print("Random Forest RMSE:   ",knn_rmse)

knn_mae = mean_absolute_error(Y_test, knn_pred)
print("Random Forest MAE :   ", knn_mae)
```

### Linear Regression Prediction

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
lr_pred = lr.predict(X_test)
```

```python
lr_mse = mean_squared_error(Y_test, lr_pred)
lr_rmse = np.sqrt(lr_mse)
print("Random Forest RMSE:   ",lr_rmse)

lr_mae = mean_absolute_error(Y_test, lr_pred)
print("Random Forest MAE :   ", lr_mae)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
etr = ExtraTreesRegressor()
etr.fit(X_train, Y_train)
```

```python
etr_pred = etr.predict(X_test)
```

```python
etr_mse = mean_squared_error(Y_test, etr_pred)
etr_rmse = np.sqrt(etr_mse)
print("Random Forest RMSE:   ",etr_rmse)

etr_mae = mean_absolute_error(Y_test, etr_pred)
print("Random Forest MAE :   ", etr_mae)
```

```python
rmse = [rf_rmse, knn_rmse, lr_rmse, etr_rmse]
index = ['rf_rmse', 'knn_rmse', 'lr_rmse', 'etr_rmse']
plt.scatter(x=index, y=rmse )
plt.ylabel("RMSE -->", fontdict={'fontsize':14,'fontweight':'bold'})
plt.xlabel("Models", fontdict={'fontsize':14,'fontweight':'bold'})
plt.show()
```
